```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

pgrr = []

for jdx in range(1,7,1):
    url = 'https://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page=' + str(jdx)
    with urlopen(url) as doc:
        html = BeautifulSoup(doc,'lxml')
        for idx in range(50):
            pgrr.append(html.find_all('a', class_ = 'tltle')[idx].text)
        
print(pgrr)
 
```

```python
series = pd.Series(pgrr)
series.head()
```

```python
df = pd.DataFrame({'회사명':series})
df.head()
```

```python
df.to_excel('KOSPI300.xlsx',index = True)
```
