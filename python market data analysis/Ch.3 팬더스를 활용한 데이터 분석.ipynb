```python
#최대 손실 낙폭(MDD, Maximum Drawdown)
#퀀트투자에서는 MDD를 낮추는 것이 더 유의하다고 할 정도로 중요한 지표
#MDD = (최저점- 최고점)/최저점

#KOSPI(참고로 코스피는 1980년 1월 4일 기준)의 MDD : 1994년 1145.66에서 1998년 277.37포인트로 이기간 MDD = -75.8%
# rolling함수 : 시리즈.rolling(윈도우 크기 [,min_periods=1])[.집계 함수()] -> 시리즈에서 윈도우 크기에 해당하는 개수만큼 데이터를 추출하여 집계함수에 해당하는 연산 실시
# 야후 파이낸스에서 KOSPI 지수 데이터를 다운로드한 뒤 rolling()함수를 이용하여 1년동안의 최대값과 최소값을 구하여 MDD를 계산하면

yf.pdr_override() #얘는 일단 먼저 써줘야 하는 듯
kospi = pdr.get_data_yahoo('^KS11', '2004-01-04')
window=252 #1년 동안의 개장일 252일
peak = kospi['Adj Close'].rolling(window, min_periods=1).max() # 윈도우는 이전 252일, 이전 데이터가 없으면 없는대로 있는데이터만 사용
drawdown=kospi['Adj Close']/peak - 1.0
max_dd = drawdown.rolling(window, min_periods=1).min()

plt.figure(figsize=(18,13))
plt.subplot(211) #2행 1열 중 1행
kospi['Close'].plot(label='KOSPI',title='KOSPI MDD', grid=True, legend=True)
plt.subplot(212) #2행 1열 중 2행
drawdown.plot(c='blue', label = 'KOSPI DD', grid = True, legend = True)
max_dd.plot(c='red', label = 'KOSPI MDD', grid = True, legend = True)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['DOW'],df['KOSPI'],marker = '.')
plt.xlabel('Dow Jones Industrial Average')
plt.ylabel('KOSPI')
plt.show()
```

```python
# Scipy 선형회귀 분석
from scipy import stats
regr = stats.linregress(df['DOW'], df['KOSPI'])
regr
```

```python
#상관계수 구하기
df.corr()
```

```python
#시리즈로 상관계수 구하기
df['DOW'].corr(df['KOSPI'])
```

```python
#결정계수 구하기
r_value = df['DOW'].corr(df['KOSPI'])
r_squared = r_value ** 2 # R2는 상관계수를 제곱해서 구함
r_squared
```

```python
#회귀선을 그리면
regr_line = f'Y = {regr.slope:.2f} * X + {regr.intercept:.2f}' #Scipy로 회귀분석이 선행되어야 함
plt.figure(figsize=(7,7))
plt.plot(df['DOW'], df['KOSPI'],'.')
plt.plot(df['DOW'], regr.slope * df['DOW'] + regr.intercept, 'r')
plt.legend(['DOW x KOSPI', regr_line])
plt.title(f'DOW x KOSPI ( R = {regr.rvalue:.2f})')
plt.xlabel('Dow Jones Industrial Average')
plt.ylabel('KOSPI')
plt.show()
```

```python
# 미국 국채와 상관관계를 파악해 보면
tlt = pdr.get_data_yahoo('TLT', '2002-7-30')
kospi2 = pdr.get_data_yahoo('^KS11', '2002-7-30')

df = pd.DataFrame({'X' : tlt['Close'], 'Y' : kospi2['Close']})
df = df.fillna(method='bfill')
df = df.fillna(method='ffill')

regr = stats.linregress(df.X, df.Y)
regr_line = f'Y = {regr.slope:.2f} * X + {regr.intercept:.2f}'

plt.figure(figsize=(7,7))
plt.plot(df.X, df.Y, 'gx')
plt.plot(df.X, regr.slope * df.X + regr.intercept, 'r')
plt.legend(['TLT x KOSPI', regr_line])
plt.title(f'TLT x KOSPI (R = {regr.rvalue:.2f})')
plt.xlabel('20+ year Treasury bond ETF')
plt.ylabel('KOSPI')
#상관계수에 따른 리스크 완화
#해리 마코위츠의 현대 포트폴리오 이론 : 상관계수가 낮은 자산에 투자하면 안정적으로 수익을 낼 수 있음
```
