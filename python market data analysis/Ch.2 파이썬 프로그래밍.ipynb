```python
# {} 기호 방식은 % 를 :로 바꾼 후 {}를 씌워서 출력
for x in crispr:
    print('{:s} : {:s}'.format(x,crispr[x]))
```

```python
#인터넷에서 이미지 가져오기(리퀘스트 라이브러리와 필로(pillow) 라이브러리, 맷플롯립 라이브러리 이용)
import requests
url = 'http://bit.ly/2JnsHnT'
r = requests.get(url, stream=True).raw #get() 함수 이용

from PIL import Image
img = Image.open(r)
img.show() # 다른 응용 프로그램을 이용해서(ex)알씨) 이미지를 보여줌
img.save('src.png')
```

```python
img.save('src.png')
print(img.get_format_mimetype) # 이미지에 대한 정보 요청

# 'with ~ as 파일객체:' 로 이미지 파일 복사
BUF_SIZE = 1024
with open('src.png', 'rb') as sf, open('dst.png', 'wb') as df: #파일을 바이너리 읽기모드(rb)로 열어서 바이너리 쓰기 모드(wb)로 저장
    while True:
        data = sf.read(BUF_SIZE)
        if not data:
            break
        df.write(data)

# SHA-256으로 파일 복사 검증하기
# SHA-2(Secure Hash Algorithm-2)는 미국 국가안보국이 설계한 해시 함수들의 집합 ; 해시는 데이터를 입력받아 고정 길이의 고유한 값으로 변환하는 것

#앞에서 다운받은 src.png 파일과 이를 바이너리 복사해서 생성한 dst.png 파일의 SHA-256 해시값을 구해보면 ( 제대로 복사가 완료되었다면, 파일 내용이 동일하기 때문에 해시값이 같아야함)

import hashlib

sha_src = hashlib.sha256() #해시 객체 생성
sha_dst = hashlib.sha256() #해시 객체 생성

with open('src.png', 'rb') as sf, open('dst.png', 'rb') as df:
    sha_src.update(sf.read())
    sha_dst.update(df.read())
    
print("src.png's hash : {}".format(sha_src.hexdigest()))
print("dst.png's hash : {}".format(sha_dst.hexdigest()))
```

```python
# 맷플롯립으로 이미지 가공하기 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

#맷플롯립은 기본적으로 png 이미지 포맷만 지원
dst_img = mpimg.imread('dst.png')
print(dst_img.shape) # 3024 x 2918 x 3 -> 컬러이미지
#대비를 위해 색상을 '의사 색상'으로 변환
print(dst_img) # 3차원 부동소수점으로 데이터가 구성된 것 확인
pseudo_img = dst_img[:,:,0] 
print('########################')
print(pseudo_img.shape) # 하나의 색깔만 이용한 이미지로 변경
print(pseudo_img) # 첫번째 채널만 슬라이싱 -> 2차원으로 데이터 구성
```

```python
#맷플롯립으로 이미지 비교하기
plt.suptitle('Image Processing', fontsize = 18)
plt.subplot(1,2,1) # 1행 2열 중 첫번째 그림 설정
plt.title('Original Image')
plt.imshow(mpimg.imread('src.png'))

plt.subplot(122) # 1행 2열 중 두번쨰 그림 설정
plt.title('Pseudocolor Image')
dst_img = mpimg.imread('dst.png')
pseudo_img = dst_img[:,:,0]
plt.imshow(pseudo_img)

plt.show()
```
