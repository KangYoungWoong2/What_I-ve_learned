```python
"""
1. 지도학습과 비지도학습
    (1) 지도 학습
        -정답을 알려주면서 진행되는 학습(데이터와함께 정답(=레이블=실제값=타깃=클래스=y값)이 항상 제공)
        -분류와 회귀

    (2)비지도 학습
        -정답이 없이 진행되는 학습, 데이터 자체에서 패턴을 찾아냄
        -군집화와 차원축소

2. 분류와 회귀
    -분류 : 데이터를 분리된 값으로 예측 ex)덥다 or 춥다
    -회귀 : 데이터를 연속된 값으로 예측 ex) 36도, 36.7도

    (1)분류
        - 이진분류: 레이블이 2개 ex)남자or여자
        - 다중분류: 레이블이 여러개
        -다중 레이블 분류 : 데이터가 입력됐을 때 두개 이상의 레이블로 분류 ex) 빨간 동그라미, 동그라미&세모

    (2)회귀
        -연속적인 수치로 예측

3.과대적합과 과소적합
    -데이터에서 충분히 특징을 잡아내지 못하면 '과소적합', 특징을 과도하게 학습하면 '과대적합'

    (1)과소적합
        -데이터에 충분한 특징이 주어지지 못했을 때
    (2)과대적합
        -필요이상의 특징이 주어져서 학습데이터에 대해서만 높은 정확성을 보임

4. 혼동행렬(confusion matrix)
    -모델의 성능을 평가할 때 사용되는 지표
    -실제값이 A, B, C, D이고 이에 대한 예측값이 A', B', C', D'일 때
           A'  B'  C'  D'
        A  9   1   0   0       -> A를 10번 보여줬을 때 9번은 맞췄지만 1번은 B'로 혼동
        B  1   20  3   0
        C  0   9   30  2
        D  0   0   1  14
     이런 식으로 행렬을 만드는 것
    
5. 머신러닝 모델의 성능평가
    (1)TP(true positive) - 맞은 것을 올바르게 예측하는 것
        - 혼동행렬의 대각선 부분
    (2) TN(true negative) - 틀린 것을 올바르게 예측하는 것
        -A가 아닌 클래스들을 A가 아니라고 예측한 모든 값 -> 위의 혼동행렬에서 1해 1열을 제외한 부분
    (3) FP(false positive) - 틀린 것을 맞다고 잘못 예측한 것
        - A가 아닌 B,C,D가 A'라고 예측한 부분 -> (2,1), (3,1), (4,1)
    (4) FN(false nagative) : 맞는 것을 틀렸다고 잘못 예측한 것
        - A 클래스에서 A'가 아닌 부분 (1,2), (1,3), (1,4)
    (5) 정확도
        - 가장 일반적인 모델 성능 평가 지표
        - TP / 전체 셀
    (6) 정밀도
        - 모델 예측값이 얼마나 정확하게 예측됐는가를 나타내는 지표 (어떤 한 클래스에 대해서 정의되는 듯)
        - TP / (TP + FP)
        - ex)       암환자     일반환자
             암환자   9            1
             일반환자 30           60 에서 암환자에 대한 정밀도는 9/39 = 23%
    (7) 재현율
        - 실제값 중 모델이 검출한 실제값의 비율 (어떤 한 클래스에 대해서 정의되는 듯)
        - TP / (TP + FN)
    (8) F1점수
        - F1점수 = 정밀도와 재현율의 조화평균 = 2 * 재현율 * 정밀도 / (재현율 + 정밀도)
        - 레이블이 불균일하게 분포된 경우 모델의 성능지표로 평가지표로 사용 가능

6. k-폴드 교차 검증
    -보통 전체 데이터셋의 70%를 학습데이터로, 20%를 테스트테이터로, 10%를 검증 데이터로 사용하는데 데이터셋이 충분하지 않으면 검증 데이터 마저 학습데이터로 써야 함
    -k-폴드 교차 검증은 이를 극복하기 위한 방법으로, 학습데이터를 n 개로 나눈 후 (1/n) 씩 총 n번의 검증을 시행 -> 각 검증에서의 정확도를 n으로 나눠 최종 정확도 계산
"""
```
