```python
import numpy as np
x = np.arange(100,300,10) #np.arange(start, end, interval), interval은 생략 가능
y = np.random.randn(20) #np.random.randn(n), n은 발생시킬 난수의 개수, 표준정규분포를 따르는 난수를 n개반환
plt.plot(x,y)
```

```python
plt.plot(y) #y축만 지정해서도 그래프 그리기 가능
```

```python
plt.figure(figsize=(10,5)) #그래프 사이즈 조절 후 그래프 그리기
plt.plot(x,y)
```

```python
#그래프를 그릴때마다 사이즈 지정하기가 귀찮으면 rc 라는 파라미터 생성
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams # rcParams를 이용해서 설정할 수 있는 파라미터를 보여줌
```

```python
plt.plot(x,y) #그래프 사이즈가 10x5에 고정된 것 확인
```

```python
#그래프 모양 설정하기
#그래프 모양은 plt.plot() 안에 속성값으로 지정, 색상은 color='색상', 선 모양은 linestyle = '선모양', 마커는 marker ='표시모양'
#색상은 일상에서 사용하는 표현으로 지정, 레드(r),블루(b),그린(g),블랙(b),마젠타(m),흰색(w),청록색(c) 등은 첫자로 지정 가능
#선 모양은 '--' : 롱대쉬, '-' : 직선, ':' : 점선, '-.' : 롱-닷 대쉬로 표현
#마커 모양은 '.' : 점, 'o' : 원, '8' : 팔각형', 's' : 사각형, 'p': 육각형, '*' :별, '+' : 더하기 등 다양함
#더 많은 마커 모양은 https://matplotlib.org/api/markers_api.html
plt.plot(x,y, color='r', linestyle='--', marker='o') #plt.plot(x,y, 'ro--')와도 같은 표현
```

```python
#이 외에도 markersize = '숫자', 마커색 markerfacecolor = '색상', 마커경계색 markeredgecolor = '색상', 투명도 alpha ='0~1사이이 소수' 등을 통해 속성 지정 가능
#제목, 축 이름, 범례삽입
#제목은 plt.title('제목')
plt.plot(x,y)
plt.title('제목') #제목이 한글일 경우 따로 설정하지 않으면 깨져서 나옴
```

```python
#그래프에서 한글사용 세팅
plt.rcParams['font.family'] = 'Malgun Gothic' #맑은 고딕
plt.rcParams['font.size'] = 12
plt.rcParams['axes.unicode_minus'] = False
#맷플롯립에서 사용할 수 있는 폰트 리스트 조회
from matplotlib import font_manager
set(sorted([f.name for f in font_manager.fontManager.ttflist]))
```

```python
plt.plot(x,y, color ='m', linestyle = ':', marker = '*', markersize = '20')
plt.title('제목')
```

```python
#그리드(점선)을 추가
plt.plot(x,y, color ='m', linestyle = ':', marker = '*', markersize = '20')
plt.title('제목')
plt.grid(True)
```

```python
#노트북 한 셀에서만 폰트 적용하는 방법
font_title = {'family' : 'Gulim', 'color' : 'black', 'weight' : 'bold' , 'size' : 20}
font_label = {'family' : 'Gulim', 'color' : 'black', 'weight' : 'regular' , 'size' : 10}
plt.plot(x,y)
plt.title('제목', fontdict = font_title)
plt.xlabel('X값(독립변수)', fontdict = font_label) #X축 지정
plt.ylabel('Y값(종속변수)', fontdict = font_label) #Y축 지정
plt.grid(True)
```

```python
#그래프 두개를 같이 나타낼 때
plt.plot(x,y, label = '그래프1')
plt.plot(x, y+1, color = 'orange', label = '그래프2')
plt.title('제목', fontdict = font_title)
plt.xlabel('X값', fontdict = font_label)
plt.ylabel('Y값', fontdict = font_label)
plt.grid(True)
plt.legend(loc = 'best') #범레 표시하는 위치는 plt.legend(loc='위치') 로 표시, 가장 좋은 위치를 컴퓨터가 표시하라고 할 땐 loc='best'
```

```python
#여러개의 그래프를 모아서 보여주기
fig = plt.figure(figsize = (16,10)) #그래프 사이즈를 지정 후 fig에 할당

ax1 = fig.add_subplot (2, 3, 1) #2행 3열 fig 중 첫째 자리를 ax1에 할당
ax1.plot(x,y)

ax2 = fig.add_subplot(2, 3, 2) #2행 3열 fig 중 둘째 자리를 ax2에 할당
ax2.hist(y) #ax2에 histogram 그리기

ax3 = fig.add_subplot(2, 3, 3) #2행 3열 fig 중 셋째 자리를 ax3에 할당
ax3.scatter(x,y)

ax4 = fig.add_subplot(2, 3, 4) #2행 3열 fig 중 넷째 자리를 ax4에 할당
ax4.plot(x,y, color = 'r' , linestyle = ':')
```

```python
#위의 과정을 다른 방법으로 해보자면
fig, axes = plt.subplots(2,3, figsize = (16,10)) 
#axes는 fig안에 들어 있는 각각의 그래프 하나씩을 가리킴 fig, axes = plt.subplots(n. m) 이라고 하면 figure 껍데기 안의 n행 m열의 subplot을 넣고 각 subplot을 axes에 할당한다는 뜻

axes[0,0].plot(y)
axes[0,1].hist(y)
axes[0,2].scatter(x,y)
axes[1,0].plot(y,'g--')
```

```python
#맷플롯립과 팬더스의 만남
import pandas as pd
""" 
N(1000,100)인 정규분포를 따르는 랜덤 주가 생성
2018년 9월 1일부터 30일 동안의 날짜로 인덱스 생성
칼럼명을 기업명으로 설정
"""

df = pd.DataFrame(np.random.randn(30,3)*100 + 1000, #평균이 1000, 표준편차가 100인  난수 생성, 30일간 3 종목이므로 np.random.randn(30,3)
                 index = pd.date_range(start='2018-09-01', periods = 30, freq = 'D'), 
                 columns = ['천리마마트', '히드라마트', '터렛마트'])
df.plot.line() #팬더스와 함께 그려낸 기본 차트
```

```python
#3D 그래프 그리기
from mpl_toolkits.mplot3d import Axes3D #맷플롯립의 추가 기능인 mplot3d 이용
fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(111, projection = '3d') # fig를 1행1열로 지정 + 3d지정
for z, n in zip([1,2,3], df.columns): #각 종목에 임의의 숫자 1,2,3,을 부여해서 이름과 묶어줌
    left = df[n].index #X축값
    height = df[n].values #Z축값
    zs = z #Y축값
    ax.bar(left, height, zs, zdir = 'y', alpha = 0.8) #3D 막대 그래프 그리기

ax.set_xlabel('날짜') #X축 이름 삽입
ax.set_ylabel('종목')
ax.set_zlabel('가격')

ax.legend(df.columns) #범례삽입
```

```python
# 입체평면을 그리려면
from matplotlib import cm #평면에 사용할 컬러맵 호출
fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection = '3d')

X = np.arange(0,20) #X축 범위 지정
Y = np.arange(-10,10) #Y축 범위 지정
X,Y = np.meshgrid(X,Y) # X, Y를 매트릭스로 변환
Z = np.sqrt(Y**2 + X) #Z값 계산

surf = ax.plot_surface(X,Y,Z, cmap = cm.summer, linewidth = 1, alpha = 0.8) #cmap = cm.summer은 색깔 지정
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
fig.colorbar(surf, shrink=0.5, aspect=5) #컬러바표시
```
